```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_76034.csv', skiprows=4)

# Show the first few rows
df.head()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
df_cleaned = df[['Country Name', '2022']].dropna()
```

```python
top_10 = df_cleaned.sort_values(by='2022', ascending=False).head(10)
```

```python
top_10.head(10)
```

```python
plt.figure(figsize=(10,5))
plt.bar(top_10['Country Name'], top_10['2022'] / 1e6)
plt.xticks(rotation=45)
plt.title('Top 10 Most Populous Countries in 2022')
plt.ylabel('Population (Millions)')
plt.tight_layout()
plt.show()
```

```python
pop_data = df['2022'].dropna()


plt.figure(figsize=(10,5))
plt.hist(pop_data / 1e6, bins=30, color='skyblue', edgecolor='black')  # in millions
plt.title('Distribution of Country Populations in 2022')
plt.xlabel('Population (Millions)')
plt.ylabel('Number of Countries')
plt.show()
```
